# Resolução de exercícios gerados pelo ChatGPT.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Exercício 1

```python
df = pd.read_csv("dados.csv")
print(df.shape)
df.head()
```

```python
df.info
```

Exercício 2

```python
print(df.isnull().sum())

df["preco_unitario"].fillna(df["preco_unitario"].mean())
df["quantidade_vendida"].fillna(df["quantidade_vendida"].mean())
```

```python
df["Nova Coluna"] = df["preco_unitario"] + df["quantidade_vendida"]
df
```

```python
df_order = df.sort_values(by = "Nova Coluna", ascending=False)
df_order
```

```python
df = df.rename(columns={"nome_produto": "produto", "preco_unitario": "preco",
                        "quantidade_vendida": "quantidade", "Nova Coluna": "preco_quantidade"})
df
```

Exercício 3

```python
filtro = df["quantidade"] > 10
df_filtrado = df[filtro]
df_filtrado = df_filtrado.sort_values(by = "quantidade", ascending=True)
df_filtrado
```

```python
df[["produto", "quantidade", "produto"]]
```

```python
# Como fazer um filtro de uma coluna de um dataframe por uma letra existente.
filtro = df["produto"].str.contains("A")
df_filtrado = df[filtro]
df_filtrado
```

```python
# Filtro a partir de query
df_filtrado = df.query("quantidade > 10 & quantidade < 30 & produto.str.contains('A')")
df_filtrado
```

Exercício 4

```python
df_filtrado = df.groupby(["produto"])["quantidade"].mean()
df_filtrado
```
